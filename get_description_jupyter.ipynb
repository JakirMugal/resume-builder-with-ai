```python
import os
from scrapy import Selector
import json
html_path = r"E:\study\GitHub\RESUME-BUILDER-WITH-AI\job_data"
meta_path = r"E:\study\GitHub\RESUME-BUILDER-WITH-AI\link_path_data.json"
job_description_path = r"E:\study\GitHub\RESUME-BUILDER-WITH-AI\job_description.json"
with open(meta_path,'r') as file:
    json_data = json.loads(file.read())
meta={}
for job_name,data in json_data.items():
    print(f'Processing {job_name}')
    if 'jobs/view' not in data['link']:
        continue
    file_path = os.path.join(html_path,data['path'])
    with open(file_path, "r", encoding="utf-8") as file:
        html_content = file.read()
    selector = Selector(text = html_content)
    job_description = '\n'.join(selector.xpath('//div[contains(@class,"description")]//text()').getall())
    meta[job_name] = {
        'path':data['path'],
        'link':data['link'],
        'description':job_description
    }

with open(job_description_path,'w') as file:
    file.write(json.dumps(meta))
```
